# Cohort Analysis: For each customer, find their first purchase date (Month/Year). Then, calculate how many customers from each "cohort" returned to buy something in subsequent months.

```python
df.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df.head()
```

```python
# Transform finds the min date for each group and maps it back to the original index
df['CohortMonth'] = df.groupby('CustomerID')['InvoiceMonth'].transform('min')
```

```python
# Calculate difference in months
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    return year, month

invoice_year, invoice_month = get_date_int(df, 'InvoiceMonth')
cohort_year, cohort_month = get_date_int(df, 'CohortMonth')

years_diff = invoice_year - cohort_year
months_diff = invoice_month - cohort_month

# Calculate total months difference
df['CohortIndex'] = years_diff * 12 + months_diff
```

```python
df.head()
```

```python
# Group by Cohort and the Index, then count unique customers
cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])['CustomerID'].nunique().reset_index()

# Pivot the table so Cohorts are rows and Months passed are columns
cohort_counts = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='CustomerID')

# Convert to percentages (Retention Rate)
cohort_sizes = cohort_counts.iloc[:, 0]
retention = cohort_counts.divide(cohort_sizes, axis=0) * 100
```

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalRevenue': 'sum'
}
)
```

```python
rfm
```

```python
# Rename columns for clarity
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalRevenue': 'Monetary'
}, inplace=True)
```

```python
rfm
```

```python
# Quick summary of the RFM distribution
print(rfm.describe())
```

```python
# Example: Dividing Recency into 4 quantiles (1 is best/most recent)
r_labels = range(4, 0, -1)
rfm['R'] = pd.qcut(rfm['Recency'], q=4, labels=r_labels)

# Example: Dividing Frequency and Monetary (4 is best)
f_labels = range(1, 5)
m_labels = range(1, 5)
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=f_labels)
rfm['M'] = pd.qcut(rfm['Monetary'], q=4, labels=m_labels)
```

# Top Selling Products by Country: Find the top 3 most sold products (by quantity) for every country

```python
df.head()
```

```python
country_product_totals = df.groupby(['Country', 'Description'])['Quantity'].sum().reset_index()
country_product_totals
```

```python
top_3_per_country = country_product_totals.groupby('Country').apply(
    lambda x: x.nlargest(3, 'Quantity')
).reset_index(drop=True)

top_3_per_country
```

```python
print(top_3_per_country[top_3_per_country['Country'].isin(['United Kingdom', 'Germany'])].sort_values(by=['Country', 'Quantity'], ascending=[True, False]))
```

```python
top_3_per_country.sort_values(by=['Country', 'Quantity'], ascending=[True, False])
```

```python
# Filter first, then sort globally by Quantity
filtered_df = top_3_per_country[top_3_per_country['Country'].isin(['United Kingdom', 'Germany'])]
print(filtered_df.sort_values(by='Quantity', ascending=False))
```

# Resampling for Growth: Resample the data to a "Weekly" frequency and calculate the Week-over-Week (WoW) growth in revenue.

```python
df_time = df.set_index('InvoiceDate')
df_time.head()
```

```python
weekly_revenue = df_time['TotalRevenue'].resample('W').sum()
weekly_revenue
```

```python
wow_growth = weekly_revenue.pct_change() * 100
wow_growth
```

```python
weekly_summary = pd.DataFrame({
    'WeeklyRevenue': weekly_revenue,
    'WoW_Growth_%': wow_growth
})

print(weekly_summary.head(10))
```

# identifying "Peak Hour" analysis to see when the most transactions happen during the day

```python
# 1. Extract hour from the InvoiceDate
df['Hour'] = df['InvoiceDate'].dt.hour

# 2. Count unique transactions (InvoiceNo) per hour
# We use nunique because one order has multiple rows (line items)
peak_hours = df.groupby('Hour')['InvoiceNo'].nunique()

# 3. Find the peak hour
best_hour = peak_hours.idxmax()
print(f"The peak hour for transactions is: {best_hour}:00")
```
